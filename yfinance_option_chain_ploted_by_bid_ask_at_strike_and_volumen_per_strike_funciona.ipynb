```python
import yfinance as yf
import matplotlib.pyplot as plt

def get_option_chain(ticker):
    stock = yf.Ticker(ticker)
    option_chain = stock.option_chain()
    return option_chain

# Get ticker from user input
ticker = input("Enter the ticker symbol: ")

# Fetch option chain data
option_chain = get_option_chain(ticker)

# Extract call and put data from option chain
calls = option_chain.calls
puts = option_chain.puts

# Plot call option data
plt.figure(figsize=(12, 6))
plt.plot(calls.lastPrice, calls.strike, label='Call Last Price')
plt.plot(calls.ask, calls.strike, label='Call Ask Price')
plt.plot(calls.bid, calls.strike, label='Call Bid Price')
plt.xlabel('Price')
plt.ylabel('Strike Price')
plt.title('Call Option Chain')
plt.legend()
plt.grid(True)
plt.show()

# Plot put option data
plt.figure(figsize=(12, 6))
plt.plot(puts.lastPrice, puts.strike, label='Put Last Price')
plt.plot(puts.ask, puts.strike, label='Put Ask Price')
plt.plot(puts.bid, puts.strike, label='Put Bid Price')
plt.xlabel('Price')
plt.ylabel('Strike Price')
plt.title('Put Option Chain')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def get_option_chain(ticker):
    stock = yf.Ticker(ticker)
    option_chain = stock.option_chain()
    return option_chain

# Get ticker from user input
ticker = input("Enter the ticker symbol: ")

# Fetch option chain data
option_chain = get_option_chain(ticker)

# Extract call and put data from option chain
calls = option_chain.calls
puts = option_chain.puts

# Plot call option data
plt.figure(figsize=(12, 6))
plt.plot(calls.lastPrice, calls.strike, label='Call Last Price')
plt.plot(calls.ask, calls.strike, label='Call Ask Price')
plt.plot(calls.bid, calls.strike, label='Call Bid Price')
plt.scatter(calls.volume, calls.strike, color='red', marker='o', label='Call Volume')
plt.xlabel('Price / Volume')
plt.ylabel('Strike Price')
plt.title('Call Option Chain')
plt.legend()
plt.grid(True, which='both', axis='both', linestyle='-', linewidth=0.5)
plt.show()

# Plot put option data
plt.figure(figsize=(12, 6))
plt.plot(puts.lastPrice, puts.strike, label='Put Last Price')
plt.plot(puts.ask, puts.strike, label='Put Ask Price')
plt.plot(puts.bid, puts.strike, label='Put Bid Price')
plt.scatter(puts.volume, puts.strike, color='red', marker='o', label='Put Volume')
plt.xlabel('Price / Volume')
plt.ylabel('Strike Price')
plt.title('Put Option Chain')
plt.legend()
plt.grid(True, which='both', axis='both', linestyle='-', linewidth=0.5)
plt.show()
```
